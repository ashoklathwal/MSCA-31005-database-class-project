```python
combined_mem_grps = pd.merge(members, groups_upd, on='group_id')
combined_mem_grps.head()
```

```python
combined_mem_grps['dummy'] = 1
```

```python
membership_crosstab = combined_mem_grps.pivot_table(values='dummy', index= 'member_id', columns='Group Name', fill_value=0)
membership_crosstab.head()
```

```python
membership_crosstab.shape
```

```python
X = membership_crosstab.T
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)

resultant_matrix = SVD.fit_transform(X)

resultant_matrix.shape
```

```python
corr_mat = np.corrcoef(resultant_matrix)
corr_mat.shape
```

```python
test_group = 'Chicago Big Data'
```

```python
group_names = membership_crosstab.columns
groups_list = list(group_names)

big_data = groups_list.index(test_group)
big_data
```

```python
corr_big_data = corr_mat[big_data]
corr_big_data.shape
```

```python
similar_groups = list(group_names[(corr_big_data<1.0) & (corr_big_data > 0.95)])
```
